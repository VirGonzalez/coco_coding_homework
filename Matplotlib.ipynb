```python
from matplotlib import pyplot as plt
import numpy as np
import math
from scipy.stats import norm
import statistics
```

```python
x = np.linspace(-3,3,100)
```

```python
plt.plot(x, 1/(sigma1 * np.sqrt(2 * np.pi)) *np.exp(-0.5)**(x / sigma1)**2 ,linewidth=2, linestyle="--", label=" sigma = 1")
plt.plot(x, 1/(sigma2 * np.sqrt(2 * np.pi)) *np.exp(-0.5)**(x / sigma2)**2,linewidth=2, linestyle="-", label="sigma = 0.5")
plt.plot(x, 1/(sigma3 * np.sqrt(2 * np.pi)) *np.exp(-0.5)**(x / sigma3)**2,linewidth=2, linestyle="dashed", label="sigma = 0.25")
plt.xlabel('x values')
plt.ylabel('g(x)')
plt.legend()
# Add the decoration | Agrega la decoracion
plt.grid()
# Add the plot title | Agrega el titulo de la gr√°fica
plt.title("Distribucion normal")
plt.show()
```

```python
from pathlib import Path
```

```python
berkeley_data_file = Path(".")/ "temperature-central-park.dat"

berkeley_data_file
```

```python
datos = np.loadtxt(berkeley_data_file)

datos
```

```python
datosC = 5/9*(datos[:, 1]-32)
datosC
```

```python
years, temperaturas = datos[:, 0], datosC
```

```python
plt.plot(years, temperaturas, "o-")

# Add the axes names | Agrega el nombre de los ejes
plt.xlabel("A√±os")
plt.ylabel("Temperatura [C]")

# Add the plot title | Agrega el titulo de la gr√°fica
plt.title("Temperatura media anual de Central Park (NY,USA)")

# Add the decoration | Agrega la decoracion
plt.grid()
plt.savefig("temperaturaCP.png", dpi=300)
plt.show()
```
